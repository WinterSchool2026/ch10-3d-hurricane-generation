```python
import numpy as np
import shutil
from pathlib import Path
import os
```

```python
# pretraining
folder = Path("/home/emiliano/Documents/3DClouds_data/")
```

```python
tot_dict = {"pretraining":{"goes":107, "himawari":95 , "msg":67}, 
           "finetune":{"goes":97, "himawari":105 , "msg":106}}
```

```python
smpl_dict = {"pretraining":{"goes":5, "himawari":4 , "msg":8}, 
           "finetune":{"goes":17, "himawari":10 , "msg":2}}
```

```python
# sample m random integrs from 0 to n-1 with nsumpy
np.random.seed(42)
np.random.choice(tot_dict["pretraining"]["goes"], smpl_dict["pretraining"]["goes"], replace=False)
```

```python
np.random.seed(42)
smpl = {ds:{sat:np.random.choice(tot_dict[ds][sat], smpl_dict[ds][sat], replace=False)  for sat in smpl_dict[ds].keys()} for ds in smpl_dict.keys()}
smpl
```

```python
files = os.listdir(folder / "tacos" / "pretraining" / "msg")
len(files)
```

```python
files
# find the ones which end in .tacozip
files = [f for f in files if f.endswith(".tacozip")]
files
```

```python
for ds in smpl.keys():
    for sat in smpl[ds].keys():
        for idx in smpl[ds][sat]:
            
            src = folder / "tacos" / ds / sat 
            dst = folder / "miniset" / ds / sat
            files = os.listdir(src)
            files = [f for f in files if f.endswith(".tacozip")]
            
            print("idx: ", idx)
            print("num files: ", len(files))
            print(f"Copying {src} to {dst}")
            
            file = files[idx]
            
            shutil.copy(src / file, dst / file)
```
